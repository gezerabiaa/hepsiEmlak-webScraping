```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.hepsiemlak.com/istanbul-kiralik"
response=requests.get(url)
html_icerik=response.content
soup=BeautifulSoup(html_icerik,"html.parser")
isim=soup.find_all("div",attrs={"class":"list-view-title"})
```

```python
fiyat=soup.find_all("span",attrs={"class":"list-view-price"})
```

```python
liste= list()
for i in range(len(isim)):
    isim[i] = (isim[i].text).strip("\n").strip()
    fiyat[i] = (fiyat[i].text).strip("\n").strip()
    liste.append(isim[i],fiyat[i])

cıktı = pd.DataFrame(liste,columns = " ürün isimleri", "ürün fiyatları")
cıktı
```
